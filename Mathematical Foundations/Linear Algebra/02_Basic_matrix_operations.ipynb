# Basic Matrix Operations
## 1) Addition
#### Matrices can be added element by element. 
#### ![alt text](equations/Matrix_addition.png)
#### The matrices should be of same size.

## 2) Subtraction
#### Matrices can be subtracted element by element. 
![alt text](equations/Matrix_subtraction.png)
#### The matrices should be of same size.

## 3) Scalar multiplication
#### We can multiply a scalar to a matrix. If we take the output of the subtraction matrix, we can see that all the elements inside the output matrix is a multiple of -4.
![alt text](equations/scalar_multiplication.png)
#### Each element is multiplied with this scalar to get the actual output.

## 4) Matrix multiplication
#### Matrices can be multiplied if the columns of one matrix equals the number of rows of the other matrix. The output size is the number of rows of first matrix and the number of columns of the second matrix.
