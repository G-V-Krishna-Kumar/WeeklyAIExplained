## 4) Matrix multiplication
#### Matrices can be multiplied if the columns of one matrix equals the number of rows of the other matrix. The output size is the number of rows of first matrix and the number of columns of the second matrix. 
#### It is performed by using the dot products of each row of A with each column of B.

![alt text](equations/matrix_multiplication.png)

#### Here, to compute the first element in the output, the dot product of Row 1 of A and Column 1 of B is calculated. Similarly, all the other elements in the output matrix (For example C where C = AB) are calculated in this way.

```python
import numpy as np


A = np.array(
    [
        [1, 2], 
        [3, 4]
    ]
)


B = np.array(
    [
        [5, 6],
        [7, 8]
    ]
)


C = np.matmul(A, B)
C
```

#### The above numpy code calculates the matrix multiplication using the matmul function.

## 7) Inverse

#### Imagine inverse as a Matrix which brings back a given tranformation back to the original form.
#### The visualization is imagine a Matrix A which is a transformation matrix which rotates left.

![alt text](<equations/left rotation.png>)

#### Here, the x basis vector moves to the direction of y basis vector. And the y basis vector moves to the (-1, 0). Hence, we will have a anti-clockwise roation of 90 degrees. 
#### If we imagine a vector **x** = [1, 1], then after we apply the transformation, we will get A**x** = [1 -1]. So this is exactly 90 degrees.
#### We can even check if they are perpendicular, if it actually rotated 90 degrees, then the dot product must be 0. 
#### $[1, 1]^{T}$ . [1, -1] = 1 * 1 + 1 * (-1) = 1-1 = 0. Hence, the vectors are perpendicular to each other.


#### Now we have this vector let's say **y** = [1 -1]. We want to know what was the original position before the transformation A is applied. In other words, what is the vector **x** which when transformed using A gives vector **y**. 
#### A**x** = **y**
#### We know A, we know y. Now we know **x** = [1 1] from above. But what if we don't know. What we can do is try to imagine another transformation which rotates right. In other words, we bring back the transformation to the original space. So, if a transformation A rotates 90 degrees anti-clockwise, the new transformation we need should rotate 90 degrees clockwise. This is known as the inverse matrix.

![alt text](<equations/right rotation.png>)

#### Now the calculation becomes $A^{-1}$**y** = [1 1] which is **x**

```python
import numpy as np

A = np.array([[0, -1],
              [1, 0]])

A_inv = np.linalg.inv(A)

print(A_inv)
```
