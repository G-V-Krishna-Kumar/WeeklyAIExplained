#### Imagine a x-y coordinate system (2 Dimension space). To represent a point, we can use a vector. This vector will consist of 2 elements because we need 2 values to represent how far the point is from each axis. 

#### For example, the point (2, 5) means the point is 2 units from the origin on the x-axis and 5 units from the origin on the y-axis.
#### Origin is the (0, 0, ...) point or the zero vector (or [0, 0] for the 2D space). We can keep this as a reference point and and then measure any other point.

![alt text](<equations/vector graph example.png>)

## Example use case of matrices and vectors:
#### Let's assume there are m equations and n unknowns. 
#### 1) The coefficients of the unknown can be represented using a matrix, A, of size m*n.
#### 2) The unknown variables can be represented using a vector, x, of size n*1.
#### 3) The set of constatns can be represented using another vector, b, of size m*1.

#### To find the vector x, we can use Ax = b. To get an intuitive understanding, we can say that given the matrix A and vector b, we need to find vector x such that Ax = b.
#### A can be thought of like a transformation matrix. Geometrically speaking, we can see a vector b, and we need to find x such that when the transformation is applied on x, we get b. This can be easily done using the inverse of A. 

#### x = $A^{-1}b$

#### We can use visualizations to get intuitive understanding of what the matrix is actually doing.
