#### For example, the set {(1, 0), (0, 1)} is not a vector space because:
#### Addition: (1, 0) + (0, 1) = (1, 1). Since, (1, 1) is not present in the set, the given set is not a vector space.

# Span
#### Span is all the possible linear combinations of the given set of vectors. For example, span{[1 0], [0 1]}, we can visualize it by thinking of all the possible linear combinations of these vectors. The possible vectors or the points we can reach is x(1, 0) + y(0, 1).

#### we can visualize this by imagining "What possible vectors we can reach from the given set of vectors by using vector addition and scalar multiplication?". For Span{[1, 0]; [0, 1]}, all the possible vectors we can reach is the 2D space. Hence, the Span is $R^2$ in this case.

# Subspace

#### Imagine a vector space V, say $R^2$. so the vector space is the 2-Dimensional space. We can choose a smaller region like for example, a line.
#### The set of vectors in the subspace already satisfies most of the conditions, the only conditions we need to check are the addition and scalar multiplication rule.
#### For example, if we take $R^2$ as a vector space, any line which passes through the origin is a subspace. This is because:
#### i) Adding any two vectors will result in a new vector which falls in the line.
#### ii) Scaling the vector will result in a new vector in the line itself. 
#### Hence, a line through origin is an example of a subspace of $R^2$

# Linear Dependence
#### Given a set of vectors, if any one of the vector is a linear combination of other, then it is linearly dependent.
#### For example, the set {(1, 1), (2, 2)} is linearly dependent because (2, 2) is just a scaled version of (1, 1). That is, we can write (2, 2) as 2(1, 1). 
#### So, if we say Span{[1, 1], [2, 2]}, it will just be a 1-Dimensional subspace of $R^2$ because the span is just a line. So it's just a line which passes through the origin and has a slope 1.
#### Even though there are 2 points, since it is linearly dependent, the output space is reduced.
#### In simple words, even though there are 2 vectors, they point in the same direction, so the vectors we can reach is the line in which the vector is pointing.

# Linear Independence

#### Given a set of vectors, if no vector can be written as a linear combination of the others, then it is linearly independent.
#### For example, the set {(1, 0), (0, 1)} is linearly independent because we cannot obtain (1, 0) from linear combinations of (0, 1). Similarly, we cannot obtain (0, 1) as the linear combination of (1, 0).
#### If we say Span{(1, 0), (0, 1)}, it will be $R^2$ space because the linear combination for the span x(1, 0) + y(0, 1) = (x, y). This is the whole $R^2$ space.
#### So, here, we had 2 vectors and the output space is the $R^2$ space.

# Basis
#### Basis is the smallest or the minimum set of linearly independent vectors needed to span the whole space.
#### For example, Span {(1, 1), (2, 2)} we seen above is just a line through the origin. So, the basis vector is {(1, 1)}.
#### Span {(1, 0), (0, 1)} is the $R^2$ space and the basis vectors are {(1, 0), (0, 1)}

# Dimension
#### It is just the number of vectors in the basis. It is written as Dim(V).
#### If basis = {[1, 1]}, then the dimension is 1.
#### If Basis = {(1, 0), (0, 1)}, then the dimension is 2.
