#### There are 3 possible scenarios. It can either have a unique solution, infinitely many solutions or no solution at all. We can visualize this by thinking the input and output spaces. If m = n = 2, it means there are 2 equations and 2 unknowns. Input space is $R^2$ and the Codomain space is also $R^2$.
#### If Rank(A) = n, the actual output space (where the output values are) has dimension 2. And there is a unique solution. There is a unique vector (x) such that when the Ax operation is performed, we get the b vector.
#### If Rank(A) < n, there are 2 cases:
#### i) When Rank(A) = Rank([A|b]) The system is consistent. There are infinite solutions. For example, when a 2D space is transformed to a line, and when the b lies on that line, there are infinite ways to reach b. In other words, when the transformation A is performed, there are infinte vectors **x** such A**x** = **b**
#### ii)  When Rank(A) < Rank([A|b]) The system is inconsistent. There is no solution. For example, when a 2D space is transformed to a line, and when the b lies outside the line, there is no way to reach b. In other words, when the transformation A is performed, there is no vector **x** such that A**x** = **b**

## Homogeneous system of equations: A**x** = **0**

#### A**x** = **0** means that after transformation A is performed, what vector or set of vectors will reach or point to the origin (**0** vector)
#### This is always consistent because **x** = **0** is always a solution as the linear transformation satisfies A**0** = **0**.

#### If Input space is $R^3$ and the dimension of the Null Space is 0, then it means that there is only 1 vector such that A**x** = **0**
#### Similarly if the dimension of the Null Space is 1, then it means that there is a line such that all the vectors on that line will give the output vector as **0** vector.

#### Nullity is the dimension of the Null Space. So, a Nullity(A) = 2 means the Null space is a 2D plane through the origin.

## Rank Nullity Theorem
#### This theorem states that Rank(A) + Nullity(A) = n
#### n = Dimension of input space

#### This can be visualized as every direction in input space is collapsed either into the output space or the null space.
#### This is useful because if some input features are redundant, a transformation can compress the data into fewer useful directions. Then Rank(A) tells how many useful directions are preserved, and Nullity⁡(A) tells how many directions are lost because they do not create any new distinction in the output.
