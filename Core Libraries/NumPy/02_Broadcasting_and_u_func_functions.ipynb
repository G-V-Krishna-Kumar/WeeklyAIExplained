# Broadcasting
#### Broadcasting changes the shape of the given numpy array such that the addition or multiplication can be performed.
#### Assume A and B are two numpy arrays. To use broadcasting, the arrays must satisfy one of the two rules:
#### i) The number of columns in A = number of rows in B
#### ii) The number of columns in A or the number of rows in B is 1

#### Example:

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 2]])

print(A.shape)
print(B.shape)
```

#### Here, it satisfies rule 2. The number of rows in B = 1 so broadcasting can be performed.

## Main u-func functions
## i) add, subtract, multiply, divide
#### Each operation is performed element wise.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

C = np.add(a, b)
D = np.subtract(a, b)
E = np.multiply(a, b)
F = np.divide(a, b)

print("add: ", C)
print("subtract: ", D)
print("multiply: ", E)
print("divide: ", F)
```

## ii) exp, log
#### exp is the exponent function. 
#### log is logarithm function.
#### Examples of where these functions used:
#### exp() is used to compute softmax.
#### log() is used to compute log-loss, cross-entropy.

```python
import numpy as np

a = np.array([0, 1, 2])
b = np.array([1, np.e, np.e**2])

exp = np.exp(a)
log = np.log(b)

print("exp: ", exp)
print("log: ", log)
```

## iii) maximum, sqrt, sum
#### maximum() can be used in ReLU activation function with maximum(0, a)
#### sqrt() can be used in distance calculations
#### sum() can be used for summing loss

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([100, -100, 1, 2, -1, 1000])

maximum = np.maximum(0, b)
sqrt = np.sqrt(a)
sum = np.sum(a)

print("maximum: ", maximum)
print("sqrt: ", sqrt)
print("sum: ", sum)
```

## iv) mean, std, var
#### mean() is used in calculating averages
#### std() is used for normalization
#### var() is used for measuring the spread of data.

```python
import numpy as np

a = np.array([1, 2, 3])

x_bar = np.mean(a)
s = np.std(a)
s_squared = np.var(a)

print("mean: ", x_bar)
print("standard deviation: ", s)
print("variance: ", s_squared)
```

## v) clip, abs, square, argmin, argmax
#### clip(array, min_value, max_value) is used to clip the the given array within the specified range [min_value, max_value]. #### abs(x) will give the positive value. If input is in the range [0, inf), it will return the value as it is since it is possible. If the input is in the range (-inf, 0), then it will give the positive value ignoring the negative sign. 

#### argmin(a) will give the index of the minimum value in the array a.
#### argmax(a) will give the index of the maximum value in the array a.
