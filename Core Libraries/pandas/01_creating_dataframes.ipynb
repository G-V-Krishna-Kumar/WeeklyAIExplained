## Series, DataFrame
#### Both are the common and most important Data Structures in pandas.
#### Series(data, index=[<Row_Labels>])
#### DataFrame({col1: Values, Col2: Values, ...})

#### Series is a 1D labeled array, so it can store 1 column of data.
#### DataFrame is a 2D labeled array, so it can store in rows and columns format.

#### Example use case for AI:
#### Series can be used as a target column

```python
import pandas as pd

y = pd.Series([10000000, 20000000, 15000000, 85000000])
y
```

#### y can be a target variable which can store the prices of a house for house prediction task.

#### Similarly, DataFrame() is used to store a set of features. For example:

```python
import pandas as pd

X = pd.DataFrame({
    "sqft": [1400, 2400, 2200, 1200],
    "bedroom": [2, 3, 3, 2],
    "bathroom": [1, 2, 2, 1]
})

X
```

## iii) head(), tail()
#### head(n) will display the first n records (or rows) of the DataFrame. Default value of n is 5.
#### tail(n) will display the last n records (or rows) of the DataFrame. Default value of n is 5.
#### Both of these are useful if we want to know how the data is stored and what the features are.

```python
import pandas as pd
import numpy as np


arr = np.arange(1, 101, 1)
df = pd.Series(arr)

print("Using head():")
print(df.head())
print()
print("Using tail():")
print(df.tail())
```

## iv) describe(), info()
#### describe() gives the important statistics data. It gives the following:
#### a) count - number of records in the dataframe
#### b) mean - mean of the column (if it is numerical)
#### c) std - standard deviation of the column
#### d) min, max - minimum and maximum value in the column
#### e) 25%, 50%, 75% - the 25th, 50th, 75th percentile of the column

#### info() gives the description of the dataframe including the column name, datatype, number of indexes (rows)

```python
import pandas as pd
import numpy as np


arr = np.arange(1, 101, 1)
df = pd.Series(arr)

print(df.describe())
print()
print("Using info():")
print(df.info())
```
